```python
from metric import Space
from metric import metrics
```

## Step 1 - Build a toy space

The records include two close spelling families and one distant outlier.

```python
records = ["cat", "cot", "coat", "dog", "dogs", "catalog"]
space = Space(records, metrics.Edit())

assert len(space) == 6
space.pairwise(ids=[0, 1, 2])
```

## Step 2 - Ask for neighbors

`count=` states the intent without choosing an index implementation.

```python
from metric.exceptions import StrategyUnavailableError

try:
    space.neighbors("cut", count=3)
    raise AssertionError("neighbors should require a native binding")
except StrategyUnavailableError as exc:
    message = str(exc)

assert "native C++ binding" in message
message
```

## Step 3 - Find groups and outliers

Default grouping and outlier scoring are deterministic for small examples.

```python
for call in (lambda: space.groups(count=2), lambda: space.outliers(count=2)):
    try:
        call()
        raise AssertionError("operation should require a native binding")
    except StrategyUnavailableError:
        pass

# The explicit distance matrix the native algorithms consume is an adapter.
matrix = space.pairwise()
assert matrix[0][0] == 0
len(matrix)
```

## Exercise

Use a radius query around `cat` and compare it with the nearest-neighbor query.

Common pitfall: radius and count answer different questions. Radius asks for all records within a threshold; count asks for the best N records.

```python
try:
    space.neighbors("cat", radius=1)
    raise AssertionError("radius search should require a native binding")
except StrategyUnavailableError as exc:
    message = str(exc)

assert "native C++ binding" in message
message
```
