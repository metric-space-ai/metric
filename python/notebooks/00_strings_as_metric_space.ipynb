# Tutorial: Strings As Metric Space

Audience:
- Python users who want to model records directly as a finite metric space.

Prerequisites:
- Basic Python lists and functions.

Learning goals:
- Build a `Space` over strings.
- Query nearest neighbors without creating vectors.
- Inspect the pairwise distance matrix.

## Outline

1. Import the public facade.
2. Build a string metric space.
3. Ask a nearest-neighbor question.
4. Inspect pairwise distances.
5. Try a small exercise.

```python
from metric import Space
from metric import metrics
```

## Step 1 - Build the space

The records stay as strings. Edit distance supplies the geometry.

```python
records = ["metric", "metrics", "matrix", "tree"]
space = Space(records, metrics.Edit())

assert space.distance(0, 1) == 1
space.ids
```

## Step 2 - Query neighbors

The query is another string; no embedding or tokenization step is required.

```python
from metric.exceptions import StrategyUnavailableError

try:
    space.neighbors("metricks", count=2)
    raise AssertionError("neighbors should require a native binding")
except StrategyUnavailableError as exc:
    message = str(exc)

assert "native C++ binding" in message
message
```

## Step 3 - Inspect distances

`pairwise()` gives the finite metric-space matrix for small spaces.

```python
matrix = space.pairwise()

assert matrix[0][0] == 0
assert matrix[0][1] == matrix[1][0]
matrix
```

## Exercise

Change the query to `metrix` and predict whether `metric` or `metrics` is closer.

Common pitfall: comparing record positions instead of record IDs. Use the returned ID to index `records` only when IDs are the default integer IDs.

```python
distances = [(records[index], space.distance(0, index)) for index in range(len(records))]

assert distances[0] == ("metric", 0)
distances
```
